```python
# English -> Emoji translator
# Example: This concert is wild -> 
# Strategy:
# Tokenize by word
# Pick emoji with high enough similarity score (Or nothing)

```

```python
import csv
emojis = []
with open('emoji_df.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',', quotechar='|')
    later = False
    for row in spamreader:
        if later:
            emojis += [row]
        later = True
```

```python
len(emojis)
```

```python
from nltk.corpus import wordnet as wn
import nltk
```

```python
emoji_description_list = []
for emoji in emojis:
    emoji_description_list += [(emoji[0], emoji[1].split())]
```

```python
# for word in input:
#     do similarity with every emoji
#     if similarity is > .024
#         pick emoji

params = {
    "minimum_similarity": 0.3
}

def score_emoji_similarity(word, emoji_labels):
    try:
        scores = [0]
        word_synset = wn.synsets(word)[0]
        for label in emoji_labels:
            similarity_score = 0
            try:
                label_synset = wn.synsets(label)[0]
                similarity_score = word_synset.path_similarity(label_synset)
            finally:
                scores += [similarity_score]
    finally:
        return max(scores)

def find_most_similar_emoji(word, emoji_list):
    emoji_scorecard = []
    for emoji in emoji_list:
        emoji_scorecard += [(emoji[0], score_emoji_similarity(word, emoji[1]))]
    emoji_scorecard = sorted(emoji_scorecard, key=lambda x: x[1], reverse=True)
    return emoji_scorecard[0]

def emojify(input, debug=False):
    output = []
    debug_info = []
    tokenizer = nltk.RegexpTokenizer(r"\w+")
    input_words = tokenizer.tokenize(input)
    for word in input_words:
        emoji, score = find_most_similar_emoji(word, emoji_description_list)
        debug_info += [(word, emoji, score)]
        if score >= params["minimum_similarity"]:
            output += [emoji]
    print(debug_info)
    return "".join(output)
        
```

```python
test_array = [
    "we go to the store",
    "wow what nice hair!"
]

for test in test_array:
    print(emojify(test))
```
